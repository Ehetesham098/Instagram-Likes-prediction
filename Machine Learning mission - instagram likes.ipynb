```python
x=data1.iloc[:,0:2] #Declare dependent variable
y=data1.iloc[:,2] #Declare the target variable
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=999)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
ypred=regressor.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)#accuracy
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,ypred)
```

```python
#Plot for residual error
plt.style.use('fivethirtyeight')
plt.scatter(regressor.predict(x_train),regressor.predict(x_train) - y_train,color = "red",s=8, label= "Train data")
plt.scatter(regressor.predict(x_test),regressor.predict(x_test) - y_test,color = "blue",s=8, label= "Test data")
plt.hlines(y=0,xmin=0,xmax=50,linewidth=2)
plt.legend( loc = 'upper right')
plt.title("Residual plot")
plt.show()
```
